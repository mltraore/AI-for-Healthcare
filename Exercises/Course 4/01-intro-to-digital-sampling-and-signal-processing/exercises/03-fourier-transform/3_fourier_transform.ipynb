# Exercise 3: Fourier Transform

You are given a signal that is composed of 5 sinusoids of varying frequency and amplitude added together. The signal is sampled at 125 Hz. Plot the Fourier transform to figure out the frequencies and the relative amplitudes of each of the sinusoids that comprise the signal. Find the frequencies in order from largest amplitude to smallest (note this will not be the same as highest frequency to lowest).

## Imports

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
import mpld3
mpld3.enable_notebook()
```

## Load Data

```python
npz = np.load("./exercise3.npz")
sig = npz['signal']
```

## Find Largest Frequency Components

```python
fs = 125
freqs = np.fft.rfftfreq(len(sig), 1/fs)
fft = np.fft.rfft(sig)
```

```python
num_top_components = 10 
top_indices = np.argsort(np.abs(fft))[-num_top_components:]

top_frequencies = freqs[top_indices]
top_magnitudes = np.abs(fft)[top_indices]
```

```python
# Plot the DFT magnitude spectrum
plt.figure(figsize=(12, 8))
plt.plot(freqs, np.abs(fft))
plt.title('Discrete Fourier Transform')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# Highlight the highest frequency components on the plot
plt.scatter(top_frequencies, top_magnitudes, color='red', marker='o', label='Top Frequencies')
plt.legend()

plt.show()
```

```python
for i in range(num_top_components):
    print(f"Top Component {i + 1}: Frequency = {top_frequencies[i]} Hz, Magnitude = {top_magnitudes[i]}")
```
